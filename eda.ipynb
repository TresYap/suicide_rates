```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv("suicide_rates_1985_2016.csv")
```

```python
data
```

```python
data.country.unique()
```

```python
data.year.describe()
```

```python
ph_data = data[data.country == "Philippines"]
```

```python
ph_data
```

```python
ph_data[ph_data.generation=="Millenials"]
```

```python
fig = px.sunburst(ph_data, path=['generation', 'year', 'age'], values='suicides_no', color='generation')
fig.show()
```

```python
ph_data.age.unique()
```

```python
agg_data = ph_data.groupby(['year', 'age']).sum().reset_index()
```

```python
fig = px.scatter(agg_data, x="year", size="gdp_per_capita ($)", y="suicides_no", color="age",
                hover_name="age", log_x=True, size_max=60, width=1000, height=800)
fig.show()
```

```python
fig = px.line(agg_data, x="year", y="suicides_no", color="age",
                hover_name="age", width=1000, height=800)
fig.show()
```

## ^ Young people in the Philippines have a higher number of suicides (probably cuz population differs)

```python
agg_data_world = data.groupby(['year', 'age', 'gdp_per_capita ($)']).sum().reset_index()
```

```python
fig = px.scatter(agg_data_world[agg_data_world.year < 2016], x="year", y="suicides_no", size="gdp_per_capita ($)", color="age",
                hover_name="age", width=1000, height=800)
fig.show()
```
